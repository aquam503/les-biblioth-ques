#### Tracée d'une ligne

* La fonction plot()

  * La fonction plot() est utilisée pour tracer des points (marqueurs)
  * Elle prend les points d'un diagramme comme paramètres.
  * Elle trace une ligne d'un point à un autre.
  * Le paramètre 1 est un tableau contenant les points sur l' axe des x .
  * Le paramètre 2 est un tableau contenant les points sur l' axe y .

Exemple:

Si nous devons tracer une ligne de (1, 3) à (8, 10),
  * nous devons passer deux tableaux [1, 8] et [3, 10] à la fonction plot.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
plt.plot(xpoints, ypoints)
plt.show()
```

1. Dessiner y en utilisant la suite 0..N-1 pour x

```python
plt.plot([3, 10])
plt.show()
```

2. Dessiner les points en les marquant

```python
"""plot(args, marker)"""
#Par un cercle
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
#Par une étoile
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = '*')
plt.show()
```

3. Modifier le style de la ligne

```python
"""plot(args, linetyle_or_ls)"""
#Ligne pointillée
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
#Ligne avec des tirets
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

##### Nuage de points
* Avec la fonction scatter()

  * Elle prend les points d'un diagramme comme paramètres.
  * Le paramètre 1 est un tableau contenant les abcisses des points .
  * Le paramètre 2 est un tableau contenant les ordonnées des points.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.show()
#Interprétation
#L'axe X indique l'âge de la voiture.
#L'axe Y indique la vitesse de la voiture lors de son passage.
#Corrélation
#Il semble que plus la voiture est récente, plus elle roule vite
```

```python
"""Deux nuages de points=Deux*scatter(x,y)"""
#Par défaut les deux nuages auront deux couleurs différentes
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.show()
#Interprétation
#'ils nous donnent tous les deux la même conclusion :
#plus la voiture est récente, plus elle roule vite.
```

```python
"""scatter(args,color)"""
#Spécifier la couleur des points avec l'argument color
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')
plt.show()
```

```python
"""scatter(args,c)"""
#Définir une couleur spécifique pour chaque point
#En utilisant un tableau de couleurs comme valeur pour l' cargument
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black",
"orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y, c=colors)
plt.show()
```

```python
"""scatter(args,c=colors, cmap='viridis')"""
#Palette de couleurs
#En utilisant l'argument mot-clé cmap avec la valeur 'viridis'
#qui est l'une des palettes de couleurs intégrées dans Matplotlib.
#elle va de 0,violet, jusqu'à 100, jaune.
#L'argument c, dans ce cas,sera un tableau avec des valeurs (de 0 à 100)
#Une valeur pour chaque point du nuage de points
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
#Paltte à inclure (optionnel)
plt.colorbar()
plt.show()
```

```python
"""scatter(args, s=sizes)"""
#Taille des points
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, s=sizes)
plt.show()
```

```python
"""scatter(args,alpha)"""
#Transparence des points avec une valeur entre 0 et 1
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
"""""""""Diagramme de batons"""""""""
#Avec la fonction bar()
"""bar(x,y) ou barh(x,y)"""
#bar(): batons tracés verticalement
#barh(): batons tracés horizontalement
#Elle prend les catégories et leurs valeurs comme paramètres.
#Le paramètre 1 est un tableau de catégories.
#Le paramètre 2 est un tableau de leur valeurs.
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
#plt.barh(x,y)
plt.show()
```

```python
"""bar(args,color) ou barh(args,color)"""
#Couleur des batons
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "red")
#plt.barh(x, y, color = "red")
plt.show()
```

```python
"""bar(args,width) ou barh(args,height)"""
#Largeur ou hauteur des batons selon le sens d'affichage
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, width = 0.1)
#plt.barh(x, y, height = 0.1)
plt.show()
```

##### Histogramme

  * Avec la fonction hist()


```python
#Histogramme représentant le nombre de valeurs dans chaque intervalle
#Générer 250 valeurs avec un écart type de 10 sur 170 
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show() 
```

```python
"""hist(x,bins)"""
#L'argument bins fait référence à chaque intervalle
x = np.random.normal(170, 10, 250)
plt.hist(x,bins = [140,150,160,170,180,190,200])
plt.show() 
```

##### Diagramme à secteurs
  * Avec la fonction pie()
     * L'argument est un tableau de valeurs
     * Pour chaque valeur du tableau, on a une pièce équivalente(coin).
     * La taille de chaque coin est: x/sum(xi)
     * le tracé du premier coin commence à partir de l'axe des x et se déplace dans le sens antihoraire

```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
#Etiquettes
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
"""pie(x, startangle)"""
#Angle de départ en degré
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
"""pie(x, explode)"""
#Coin à démarquer
#Argument sous forme d'un tableau montrant la distance de chaque coin/centre
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
"""pie(x, shadow=True)"""
#Ajoutez une ombre au camembert en définissant le shadowsparamètre surTrue
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```


"""pie(x, colors)"""
#Couleur pour chaque coin
#hormis les couleurs nomibales ou hexadécimales
#Vous pouvez utilisez des raccourcis de couleurs
#'r'- Rouge
#'g'- Vert
#'b'- Bleu
#'c'- Cyan
#'m'- Magenta
#'y'- Jaune

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python

"""legend()"""
#Légende
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 

"""legend(title)"""
#Légende avec titre
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
